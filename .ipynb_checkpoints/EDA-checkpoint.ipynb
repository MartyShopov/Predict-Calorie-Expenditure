```python
import seaborn as sns

def plot_correlation_matrix(df: pd.DataFrame, figsize=(10, 8)) -> None:
    """
    Plots a correlation matrix for the features in a pandas DataFrame.

    Parameters:
    - df: pd.DataFrame - Input DataFrame containing numeric features.
    - figsize: tuple - Size of the figure (width, height). Default is (10, 8).
    """
    # Compute correlation matrix
    corr_matrix = df.corr(numeric_only=True)
    
    # Set up the matplotlib figure
    plt.figure(figsize=figsize)
    
    # Plot the heatmap
    sns.heatmap(
        corr_matrix,
        annot=True,           # Show correlation values
        cmap="Blues",         # Use the "Blues" colormap
        fmt=".2f",            # Format annotations to 2 decimal places
        linewidths=0.5,       # Add lines between cells
        cbar=True             # Show color bar
    )
    
    plt.title("Correlation Matrix", fontsize=16)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
plot_correlation_matrix(data)
```

```python
plot_correlation_matrix(data[data.sex == "male"])
```

```python
plot_correlation_matrix(data[data.sex == "female"])
```
